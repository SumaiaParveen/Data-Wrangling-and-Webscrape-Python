# 4 Ways To Filter A Pandas Dataframe

```python
import pandas as pd

df = pd.read_csv("china_gdp.csv")
df.head()
```

## Filter Rows Whereâ€¦

```python
df[df['Value'] > 7.918412e+12]
```

```python
filt = (df['Value'] > 8.518412e+12) & (df['Value'] < 7.918412e+13)
df[filt]
```

## First/Last Rows

```python
df[:2]
```

```python
df[-2:]
```

## Where Value Is In Specified List

```python
numbers = [2012, 2013, 2014]

df[df['Year'].isin(numbers)]
```

## Where Value Is/Not null(NaN)

```python
df[df['Year'].notnull()]
```

```python
df[df['Year'].isnull()]
```
